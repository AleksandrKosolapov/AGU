```python
def reduce_mem_usage(df):
    """ iterate through all the columns of a dataframe and modify the data type
        to reduce memory usage.        
    """
    start_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage of dataframe is {:.2f} MB'.format(start_mem))
    
    for col in df.columns:
        if col not in ['url', 'referrer_num']:
            col_type = df[col].dtype

            if col_type != object:
                c_min = df[col].min()
                c_max = df[col].max()
                if str(col_type)[:3] == 'int':
                    if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                        df[col] = df[col].astype(np.int8)
                    elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                        df[col] = df[col].astype(np.int16)
                    elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                        df[col] = df[col].astype(np.int32)
                    elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                        df[col] = df[col].astype(np.int64)  
                else:
                    if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                        df[col] = df[col].astype(np.float16)
                    elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                        df[col] = df[col].astype(np.float32)
                    else:
                        df[col] = df[col].astype(np.float64)
            else:
                pass

    end_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage after optimization is: {:.2f} MB'.format(end_mem))
    print('Decreased by {:.1f}%'.format(100 * (start_mem - end_mem) / start_mem))
    
    return df
reduce_mem_usage(df)
```

```python
train = df[~df['Статус'].isna()]
test = df[df['Статус'].isna()]
del test['Статус']

train['Статус'].value_counts(dropna=False)
```

```python
from catboost import CatBoostClassifier,Pool


cats = [c for c in cats if c in train.columns]

folds = 6
c=0

for r in range(200,205):
    train = train.sample(frac=1,random_state=r).sort_values(['Статус','Основания','Код_группы']).reset_index(drop=True)
    
    for f in list(range(folds)):
        
        test_x = train[f::folds]
        y2 = test_x['Статус']
        cat_features = [test_x.columns.get_loc(col) for col in cats]
        text_features = [test_x.columns.get_loc(col) for col in text]
        del test_x['Статус']

        train_x = train[~train.index.isin(list(test_x.index))] 
        y = train.loc[~train.index.isin(list(test_x.index)),'Статус']
        del train_x['Статус']
        train_pool = Pool(data=train_x, label=y.values, cat_features = cat_features, text_features = text_features)
        test_pool = Pool(data=test_x, label=y2.values, cat_features = cat_features, text_features = text_features)

        model = CatBoostClassifier(
        loss_function='MultiClass',
        objective = 'MultiClass',
        eval_metric='TotalF1',
        learning_rate = 0.1,
        max_depth=8,
        verbose =500,
              iterations = 20000, #5000
              task_type = 'GPU',
              min_data_in_leaf = 15)

        model.fit(train_pool,plot=True,eval_set=test_pool,use_best_model=True, early_stopping_rounds=200)
        if c == 0:
            preds = model.predict_proba(Pool(data=test, cat_features = cat_features, text_features = text_features))
            c+=1
        else:
            preds += model.predict_proba(Pool(data=test, cat_features = cat_features, text_features = text_features))
```

```python
pd.DataFrame({'feature_importance': model.get_feature_importance(train_pool),
              'feature_names': train_x.columns}).sort_values(by=['feature_importance'], ascending=False)
```

```python
dfpred = pd.DataFrame(preds)
dfpred.columns = [-1,3,4]

def returncolname(row, colnames):
    return colnames[np.argmax(row.values)]

dfpred['colmax'] = dfpred.apply(lambda x: returncolname(x, dfpred.columns), axis=1)
dfpred
```

```python
import pandas as pd
import numpy as np
from nltk import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer


train = df[~df['Статус'].isna()]
test = df[df['Статус'].isna()]
del test['Статус']

dft = df['Регион_ПП'].astype(str) + ' ' + df['Город_ПП'].astype(str)

dft = dft.str.replace('[^a-zA-Zа-яА-Я09]', ' ').apply(lambda x: str(x).replace('  ',' ')).str.lower()

#texts = np.array(df['text']) #text contents in dataframe to array for processing
#vocab_length = len(nltk.word_tokenize(list(itertools.chain.from_iterable(texts))) #concatenate all the texts and tokenize the whole corpus
vect = TfidfVectorizer(ngram_range = (1,3), min_df=20) #make Tfidf Vectorizer
df = df.merge(pd.DataFrame(vect.fit_transform(dft.astype(str)).toarray(),columns=vect.get_feature_names()),left_index=True,right_index=True,how='left') #encode the text
#df['tfidf'] = list(tfidf_encodings.toarray())

dft = df['Основания'].astype(str) + ' ' + df['Уч_Заведение'].astype(str) + ' ' + df['Где_Находится_УЗ']
dft = dft.str.replace('[^a-zA-Zа-яА-Я09]', ' ').apply(lambda x: str(x).replace('  ',' ')).str.lower()
#texts = np.array(df['text']) #text contents in dataframe to array for processing
#vocab_length = len(nltk.word_tokenize(list(itertools.chain.from_iterable(texts))) #concatenate all the texts and tokenize the whole corpus
vect = TfidfVectorizer(ngram_range = (1,3), min_df=20) #make Tfidf Vectorizer
df = df.merge(pd.DataFrame(vect.fit_transform(dft.astype(str)).toarray(),columns=vect.get_feature_names()),left_index=True,right_index=True,how='left') #encode the text

reduce_mem_usage(df)

train = df[~df['Статус'].isna()].drop(text,axis=1)
test = df[df['Статус'].isna()].drop(text,axis=1)
del test['Статус']
```

```python
#LIGHT GBM

import lightgbm as lgb
params = {
          "objective" : "multiclass",
          "num_class" : 3,
          "num_leaves" : 15,
          "max_depth": -1,
          "learning_rate" : 0.001,
          "bagging_fraction" : 0.9,  # subsample
          "feature_fraction" : 0.9,  # colsample_bytree
          "bagging_freq" : 5,        # subsample_freq
          "verbosity" : -1}

cats = [c for c in cats if c in train.columns]

folds = 6
c=0

for r in range(200,220):
    train = train.sample(frac=1,random_state=r).sort_values(['Статус','Основания','Код_группы']).reset_index(drop=True)
    
    for f in list(range(folds)):
        
        test_x = train[f::folds]
        y2 = test_x['Статус'].map({4:2,3:1,-1:0})
        del test_x['Статус']

        train_x = train[~train.index.isin(list(test_x.index))] 
        y = train.loc[~train.index.isin(list(test_x.index)),'Статус'].map({4:2,3:1,-1:0})
        del train_x['Статус']

        lgtrain, lgval = lgb.Dataset(train_x, y,categorical_feature=cats), lgb.Dataset(test_x, y2,categorical_feature=cats)
        lgbmodel = lgb.train(params, lgtrain, 50000, valid_sets=[lgtrain, lgval], early_stopping_rounds=5000, verbose_eval=5000)
        
        if c == 0:
            lgpreds = lgbmodel.predict(test, num_iteration=lgbmodel.best_iteration)
            c+=1
        else:
            lgpreds += lgbmodel.predict(test, num_iteration=lgbmodel.best_iteration)
```

```python
lgpred = pd.DataFrame(lgpreds)
lgpred.columns = [-1,3,4]

def returncolname(row, colnames):
    return colnames[np.argmax(row.values)]

lgpred['colmax'] = lgpred.apply(lambda x: returncolname(x, lgpred.columns), axis=1)
lgpred
```
